# Confidence related metrics from the mouse trajectory

3 types of trajectory metrics:
1) interclick time: correct vs. incorrect
2) selection speed: correct vs. incorrect
3) dwell time:      correct vs. incorrect

```python
trajMat = loadmat(CLdata_folder + 'trajCell_allConds_YAOA.mat') # 'trajCell_allConds'
traj_lmm_group = trajMat['trajCell_allConds']
traj_lmm_pd    = pd.DataFrame(traj_lmm_group)
traj_lmm_pd
```

```python
column_name = {0: 'curriculumCond', 1: 'groupNo', 2: 'subjNo', 3: 'trialNo', 4: 'retrievalType', 5: 'resp', 6: 'trajMetLabel', 7: 'trajMet'} 
traj_lmm_pd = traj_lmm_pd.rename(columns = column_name) 
traj_lmm_pd.curriculumCond = traj_lmm_pd.curriculumCond.astype('int64')
traj_lmm_pd.groupNo        = traj_lmm_pd.groupNo.astype('int64')
traj_lmm_pd.subjNo         = traj_lmm_pd.subjNo.astype('int64')
traj_lmm_pd.trialNo        = traj_lmm_pd.trialNo.astype('int64')
traj_lmm_pd.retrievalType  = traj_lmm_pd.retrievalType.astype('int64')
traj_lmm_pd.resp           = traj_lmm_pd.resp.astype('int64')
traj_lmm_pd.trajMetLabel   = traj_lmm_pd.trajMetLabel.astype('int64')
traj_lmm_pd
```

```python
nConds   = 3 # {'interleaved', 'objectBlocked', 'locationBlocked'}
nGroup   = 2 # younger and older groups
nMeas    = 3 # three types of retrievals: object, location and full
nResp    = 2 # responses: 1-correct, 0-incorrect
nMetrics = 3 # trajectory-based confidence metrics: 1) interclick time; 2) selection speed; 3) dwell time
condsList   = ['0interleaved', '1objectBlocked', '2locationBlocked']
groupList   = ['0younger', '1older']
measList    = ['0object', '1location', '2full']
respList    = ['0incorrect', '1correct']
metricsList = ['interclickTime', 'selectionSpeed', 'dwellTime']
for iConds in range(0, nConds): # 1-interleaved, 2-objectBlocked, 3-locationBlockd
    traj_lmm_pd.loc[(traj_lmm_pd['curriculumCond'] == iConds + 1), 'curriculumCond'] = condsList[iConds]
for iGrp in range(0, nGroup): # 1-YA, 2-OA
    traj_lmm_pd.loc[(traj_lmm_pd['groupNo'] == iGrp + 1), 'groupNo'] = groupList[iGrp]  
for iM in range(0, nMeas): # 1-object, 2-location, 3-full
    traj_lmm_pd.loc[(traj_lmm_pd['retrievalType'] == iM + 1), 'retrievalType'] = measList[iM]
for iR in range(0, nResp): # 0-incorrect, 1-correct
    traj_lmm_pd.loc[(traj_lmm_pd['resp'] == iR), 'resp'] = respList[iR]
for iMt in range(0, nMetrics): # 1-interclick time; 2-selection speed; 3-dwell time
    traj_lmm_pd.loc[(traj_lmm_pd['trajMetLabel'] == iMt + 1), 'trajMetLabel'] = metricsList[iMt]
    
traj_lmm_pd
```

```python
# normalize the trial numbers 
traj_lmm_pd['trialNo'] = np.log(traj_lmm_pd['trialNo'])
traj_lmm_pd
```

```python
path_or_buf = CLdata_folder + 'trajCell_allConds_YAOA.csv' # save in the folder of online data
traj_lmm_pd.to_csv(path_or_buf = path_or_buf)
```
