### YA and OA together

```r
resp_glmm_pd_YAOA = resp_glmm_pd
```

```r
resp_glmm_pd_YAOA$trialNo = scale(resp_glmm_pd_YAOA$trialNo, center=TRUE, scale=FALSE) #scale=FALSE
resp_glmm_pd_YAOA
```

```r
choiceModel <- mixed(resp ~ trialNo*measNo*groupNo + (1 | subjNo), data = resp_glmm_pd_YAOA, family = binomial, method="LRT")
choiceModel
```

```r
# Effect size: partial R²β per fixed effect (Jaeger et al., 2017)
r2beta(choiceModel$full_model, method = 'nsj', partial = TRUE)
```

```r
# Hedges' g for age group comparison (YA vs OA)
# Aggregate to subject level before computing effect size
subj_mean <- resp_glmm_pd_YAOA %>%
  group_by(subjNo, groupNo) %>%
  summarise(prop_resp = mean(resp), .groups = 'drop')
hedges_g(prop_resp ~ groupNo, data = subj_mean)
```

```r
emm1 = emmeans(choiceModel, specs = ~ trialNo*measNo*groupNo) # same as: emmeans(choiceModel, ~ networkName) & emmeans(choiceModel, c("networkName"))
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "measNo", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```

```r
# Effect size for GLMM contrasts: log-odds ratio converted to Cohen's d
# Approximation: d = log-odds / (pi / sqrt(3))
contrast_df <- as.data.frame(summary(emm_pairs))
contrast_df$cohens_d <- contrast_df$estimate / (pi / sqrt(3))
contrast_df[, c('contrast', 'cohens_d')]
```

```r
traj_lmm_pd$trialNo = scale(traj_lmm_pd$trialNo, center=TRUE, scale=FALSE) #scale=FALSE
traj_lmm_pd
```

## Interclick time as the response variable

```r
# column_name = {0: 'curriculumCond', 1: 'groupNo', 2: 'subjNo', 3: 'trialNo', 4: 'retrievalType', 5: 'resp', 6: 'trajMetLabel', 7: 'trajMet'} 
# measList    = ['0object', '1location', '2full']
# metricsList = ['interclickTime', 'selectionSpeed', 'dwellTime']
interclickTime_pd <- traj_lmm_pd[
    traj_lmm_pd$trajMetLabel == 'interclickTime' &
    (traj_lmm_pd$retrievalType == '0object' | traj_lmm_pd$retrievalType == '1location'),
]
```

```r
interclickTime_pd
```

```r
trajModel <- lmer(trajMet ~ curriculumCond * groupNo * retrievalType * resp + (1 + retrievalType * resp | subjNo), data = interclickTime_pd)  
```

```r
# summary(trajModel)
```

```r
anova(trajModel, ddf='Satterthwaite')
```

```r
# ------ The original method for simple effect ------
# The result should be exactly the same as the below one with the conditional statement
# emm1 = emmeans(trajModel, specs = ~ curriculumCond * groupNo * retrievalType * resp, lmer.df = "satterthwaite") # specs = ~ groupNo*measNo; same as: emmeans(choiceModel, ~ networkName) & emmeans(choiceModel, c("networkName"))
# emm_pairs <- pairs(emm1, simple = "curriculumCond", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
```

```r
emm_curriculum <- emmeans(
  trajModel,
  ~ curriculumCond | groupNo * retrievalType * resp,
  lmer.df = "satterthwaite"
)
emm_curriculum
```

```r
emm_pairs <- pairs(
  emm_curriculum,
  adjust = "bonferroni"
)
emm_pairs
```

```r
summary(emm_pairs, infer = c(TRUE, TRUE))
```
